### Problem Statement
An image shows a pair of earrings. A camera is mounted with the optical axis perpendicular to the imaging plane.

**Given:**
- Focal length ($f$) = 8 mm
- Pixel size = 2.2 µm × 2.2 µm (0.0022 mm)
- Distance from lens to object plane ($d$) = 720 mm

**Goal:** Find the physical size of the earrings.

### Explanation & Formulas
The magnification $M$ relates the image size ($I$) to the object size ($O$):

$$M = \frac{f}{d - f}$$

Once magnification is known, the physical size of an object in the real world is determined by:

$$O = \frac{I}{M}$$

Since $I$ (image size in mm) is the number of pixels multiplied by the pixel size, we can define a scale factor:

$$\text{Real world size per pixel} = \frac{\text{Pixel Size}}{M}$$

```python
import PIL.Image
import matplotlib.pyplot as plt

# Load and display the image
img = PIL.Image.open('/content/earrings.jpg')
plt.figure(figsize=(10, 6))
plt.imshow(img)
plt.title('Earrings Image')
plt.axis('on')
plt.show()

# Constants from the problem
focal_length_mm = 8.0
distance_to_plane_mm = 720.0
pixel_size_mm = 0.0022

# Calculate Magnification
# M = f / (d - f)
magnification = focal_length_mm / (distance_to_plane_mm - focal_length_mm)

print(f'Magnification: {magnification:.6f}')
print(f'Scale: 1 pixel in image = {pixel_size_mm / magnification:.4f} mm in real world')
```

### Results and Interpretation
Based on the calculation above:

1. **Magnification ($M$):** 0.011236. This means the image on the sensor is about 1.12% of the actual size of the earrings.
2. **Scale Factor:** 1 pixel = **0.1958 mm**.

#### How to find the physical size:
To determine the actual size of the earrings in millimeters, measure the object in the image (in pixels) and use the following formula:

$$\text{Physical Size (mm)} = \text{Size in Pixels} \times 0.1958$$

For example, if an earring measures 150 pixels in height in the image:
$$150 \text{ pixels} \times 0.1958 \text{ mm/pixel} \approx 29.37 \text{ mm}$$
